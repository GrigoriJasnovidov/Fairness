```python
import os
import sys
sys.path.append(os.path.dirname(os.path.dirname(os.getcwd())))
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from lightgbm import LGBMClassifier

from fmclp.synthetic_dataset import synthetic_dataset
from fmclp.fmclp_algorithm import fmclp
from fmclp.write_results import general_results_write
```

```python
d = synthetic_dataset(200, influence=False)
d
```

```python
y = d.drop('target', axis=1)
x = d['target']
y_train, y_test, x_train, x_test = train_test_split(y, x)
lg = LGBMClassifier(verbose=-1)
lg.fit(y_train, x_train)
lg_preds = lg.predict(y_test)
accuracy_score(lg_preds, x_test)
```

```python
%%time
results = []

for i in range(100):
    d = synthetic_dataset(200, influence=False)
    cl = LGBMClassifier(verbose=-1)
    y = d.drop('target',axis=1)
    x = d['target']
    y_train, y_test, x_train, x_test = train_test_split(y, x)
    cl.fit(y_train, x_train)

    res = fmclp(dataset=synthetic_dataset(10000, influence=False), 
                estimator=cl, 
                number_iterations=10, 
                prefit=True, 
                interior_classifier='knn',
                verbose=False, 
                multiplier=10)
    results.append(res)  
    print(i+1)
    
general_results_write(results=results,
                      dataset_name='synthetic_independent',
                      number_experiments=100,
                      number_iterations=10,
                      save_folder='../../results/main_testing_results' ,
                      write_benefit=False)
```
